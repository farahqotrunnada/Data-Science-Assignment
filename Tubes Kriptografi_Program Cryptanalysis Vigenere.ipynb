# 2. Enkripsi

```python
# fungsi untuk melakukan enkripsi    
def encrypt(x,k):
    x = x.upper()
    k = k.upper()
    
    y = ""

    lenx = len(x)
    lenk = len(k)

    for i in range(lenx):
        j = i % lenk
        loc = (ord(x[i]) + ord(k[j])) % 26
        y += chr(loc + 65)

    return y
```

```python
# fungsi untuk melakukan dekripsi
def decrypt(y, k):
    y = y.upper()
    k = k.upper()
    
    x = ""

    leny = len(y)
    lenk = len(k)

    for i in range(leny):
        j = i % lenk
        loc = (ord(y[i]) - ord(k[j])) % 26
        x += chr(loc + 65)

    return x
```

# 4. Kriptaanalisis

```python
# fungsi untuk mengatur teks string menjadi array ukuran m x (n/m)
def arrange_text(text, m):
    array_text = text
    col = np.int(np.ceil(len(array_text)/m))
    r = m - (len(array_text) % m)

    array_text += " "*r

    y = []
    for i in range(col):
        j = i * m
        y.append(list(array_text[j:j+m]))

    return np.array(y)

# fungsi untuk menghitung index coincidence
def ic(cipher_):
    N            = len(cipher_)
    freqs        = collections.Counter(cipher_)
    alphabet     = map(chr, range(ord('A'), ord('Z') + 1))
    freqsum      = 0

    for letter in alphabet:
        freqsum += freqs[letter] * (freqs[letter] - 1)

    return freqsum / (N * (N - 1))

# fungsi untuk print index coincidence untuk setiap sub string
def print_ic(cipher_):
    m = cipher_.shape[1]
    for i in range(m):
        subtext = "".join(cipher_[:,i]).strip()
        print(ic(subtext))
  

# fungsi untuk menghitung rata-rata index coincidence
def avg_ic(cipher_):
    m = cipher_.shape[1]
    avg = 0
    for i in range(m):
        subtext = "".join(cipher_[:,i]).strip()
        avg += ic(subtext)

    return avg / m

# fungsi untuk menghitung mutual index coincidence
def mic(subtext, g):
    N            = len(subtext)
    freqs        = collections.Counter(subtext)
    alphabet     = map(chr, range(ord('A'), ord('Z') + 1))
    freqsum      = 0

    for letter in alphabet:
        letter2 = chr(((ord(letter)-65+g) % 26) + 65)
        freqsum += prob_[letter] * freqs[letter2]

    return freqsum / N

# fungsi untuk mencari gcd
def findgcd(x, y):
   while(y):
      x, y = y, x % y
   return x
```

```python
#Kasiski Test

message=text
seqSpacings = {} # keys adalah text sequences, values adalah list dari index sequences
for seqLen in range(3, 7):
    for seqStart in range(len(message) - seqLen):
        # Menentukan sequence dan menyimpannya pada seq
        seq = message[seqStart:seqStart + seqLen]

        if seq not in seqSpacings:
          # Cek sequence ini pada message selanjutnya
          for i in range(seqStart + seqLen, len(message) - seqLen):
              if message[i:i + seqLen] == seq:
                  # Menemukan sequence yang berulang.
                  if seq not in seqSpacings:
                      seqSpacings[seq] = [] # inisialisasi list
                      seqSpacings[seq].append(seqStart+1)
                  # Memasukkan index posisi
                  seqSpacings[seq].append(i+1 - seqStart)
seqSpacings
```

```python
# Mencari GCD dari proses teks kasiski
GCD=[]
for i in range(len(seqSpacings)):
    list(seqSpacings.values())[i].sort()
    x=list(seqSpacings.values())[i]
    print(x)
    k=x[0]
    xx=x[1:]
    xx=[i-k for i in x]
    num1=xx[0]
    num2=xx[1]
    gcd=findgcd(num1,num2)
    for i in range(2,len(xx)):
       gcd=findgcd(gcd,xx[i])
    print('GCD bernilai ',gcd,'\n')
    if gcd<=180:
      GCD.append(gcd)
```

```python
# Menghitung rata-rata IC tiap panjang kunci(m) dengan m yang berasal dari hasil teks kasiski serta angka 1-11
k=[i for i in range(1,11)]
L=list(set(GCD+k))
L.sort()

average_ic=[]
ave={}
for i in L:
    cipher_ = arrange_text(text, i)
    print('IC for m=',i)
    print_ic(cipher_) 
    print('Average IC :', avg_ic(cipher_),'\n')
    average_ic.append(avg_ic(cipher_))
    ave[i]=avg_ic(cipher_)
```

```python
# Mencari m(panjang kunci) yang menghasilkan average IC yang paling mendekati 0.0785163581027618

min=100
for i,j in ave.items():
  j=abs(j-sum_ic)
  if j<min:
    min=j
    k=i
k
```

```python
# Mencari kunci yang paling mungkin

key_length = k
cipher_ = arrange_text(text, key_length)
mic_matrix = np.zeros((key_length, 26))
for i in range(key_length):
    subtext = "".join(cipher_[:,i]).strip()
    for g in range(26):
        mic_matrix[i, g] =  mic(subtext, g)

key_text = ""
for i in range(key_length):
    key_text += chr(np.argmin(abs(mic_matrix[i]-sum_ic)) + 65)

print(key_text)
```
